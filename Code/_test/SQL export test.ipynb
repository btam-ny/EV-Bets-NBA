```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL
import pypyodbc
import pandas as pd
```

```python
SERVER_NAME = 'SKYNET'
DATABASE_NAME = 'test'
TABLE_NAME = 'nba_players'
```

```python
connection_string = f"""
    DRIVER={{SQL Server}};
    SERVER={SERVER_NAME};
    DATABASE={DATABASE_NAME};
    Trusted_Connection=yes;
"""
```

```python
connection_url = URL.create('mssql+pyodbc', query={'odbc_connect': connection_string})
engine = create_engine(connection_url, module=pypyodbc)
```

```python
excel_file = pd.read_csv(r'C:\Users\Brian\Main Folder\EV Bets\data\stats_full_data\data_stats_2024-03-13.csv')
excel_file.head()
```

```python
#excel_file = excel_file.rename(columns={'Unnamed: 0': 'Column 0'})
excel_file = excel_file[['First Last Name']]
excel_file.head()
```

```python
for sheet_name, df_data in excel_file.items():
    print(f'Loading worksheet {sheet_name}...')
    # {'fail', 'replace', 'append'}
    df_data.to_sql(TABLE_NAME, engine, if_exists='append', index=False)
```
